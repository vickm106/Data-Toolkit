1. What is NumPy, and why is it widely used in Python?

- NumPy, short for Numerical Python, is a fundamental library in Python that provides powerful tools for numerical computing. It's widely used for a variety of reasons:   

 i. Efficient Multi-Dimensional Arrays: NumPy's core feature is the ndarray, a homogeneous n-dimensional array object. These arrays allow you to store and manipulate large amounts of numerical data efficiently, especially compared to Python's built-in lists.   

 ii. High-Performance Operations: NumPy provides a vast collection of functions for performing mathematical operations on arrays. These functions are often implemented in highly optimized C code, making them significantly faster than equivalent Python code. This speed is crucial for scientific computing and data analysis.  

 iii. Foundation for Scientific Computing: NumPy serves as the foundation for many other scientific computing libraries in Python, such as Pandas, SciPy, and Scikit-learn. These libraries build upon NumPy's array structures and efficient operations, making it an essential tool for data science, machine learning, and other scientific fields.   

 iv. Ease of Use: NumPy provides a convenient and intuitive way to work with arrays. Its syntax is similar to that of mathematical notation, making it easy to express complex operations concisely.

 v. Open Source and Community-Driven: NumPy is an open-source library with a large and active community. This means it's constantly being improved and updated, and there's a wealth of resources available for learning and troubleshooting.

 4. Explain the use of the groupby() method in Pandas.

 - The groupby() method in Pandas is a powerful tool for data manipulation and analysis. It allows you to split a DataFrame into groups based on the values in one or more columns, and then apply functions to each group independently. This is often referred to as the "split-apply-combine" strategy.

 Here's a breakdown of how it works and why it's useful:

 i. Splitting:

 The groupby() method takes one or more column names as arguments.
It divides the DataFrame into groups based on the unique values in those columns.
For example, if you have a DataFrame with a "city" column, df.groupby('city') would group the data by unique city names.  
 ii. Applying:

 After grouping, you can apply various functions to each group.
Common functions include:
Aggregation functions: Calculate summary statistics like sum(), mean(), count(), min(), max(), etc.
Transformation functions: Modify the values within each group.
Filtering functions: Select or exclude groups based on certain conditions.  
 iii. Combining:

 Finally, the results of the applied functions are combined into a new DataFrame or Series.
Why is groupby() useful?

 Data aggregation: It allows you to easily calculate summary statistics for different groups within your data. For example, you can find the average sales per city, the total number of customers in each region, etc.
Data transformation: You can modify data within each group independently. For instance, you might normalize data within each group, or apply a different calculation to each group.
Data filtering: You can filter groups based on certain criteria. For example, you might want to analyze only the groups with sales above a certain threshold.
Data analysis: groupby() is a fundamental tool for exploring and understanding your data. It enables you to identify patterns, trends, and relationships within different segments of your data.

17. Explain the difference between apply() and map() in Pandas.

- Both apply() and map() are used in Pandas to apply functions to Series or DataFrames, but they operate at different levels and have distinct use cases. Here's a breakdown of their differences:

 map():

 Works on: Pandas Series only.  
Element-wise operation: Applies a function to each individual element of the Series.  
Function input: The function passed to map() takes a single value (an element from the Series) as input.  
Return value: Returns a new Series with the same index as the original Series, where each element has been transformed by the function.  
Use cases: Primarily used for simple element-wise transformations, like changing data types, replacing values, or applying mathematical functions to each element.  
apply():

 Works on: Pandas Series or DataFrames.  
Operates along an axis: Applies a function along a specified axis (rows or columns) of a DataFrame or to the entire Series.  
Function input: The function passed to apply() can take a Series (when applied to a DataFrame's row or column) or a single value (when applied to a Series) as input.  
Return value: Can return a Series, a DataFrame, or a single value, depending on the function and whether it's applied to a Series or DataFrame. This makes apply() more versatile.  
Use cases: Used for more complex transformations that might involve multiple elements, calculations across rows/columns, or aggregations. Also useful when the function's output depends on the structure of the row or column.

18. What are some advanced features of NumPy?

- NumPy offers a wide range of advanced features that extend its core functionality beyond basic array manipulation. Here are some notable examples:   

 i. Structured Arrays:

 Structured arrays allow you to create arrays where each element is a record or struct, containing fields of different data types. This is useful for representing complex data structures, like database records or scientific measurements with multiple properties.  
 ii. Masked Arrays:

 Masked arrays provide a way to handle missing or invalid data. You can "mask" certain elements of an array, indicating that they should be ignored during calculations. This is useful for dealing with datasets with incomplete or unreliable information.     
 iii. Memory Mapping:

 Memory mapping allows you to treat a file on disk as if it were an array in memory. This is particularly useful for working with very large files that don't fit into memory, as it allows you to access and process the data in chunks without loading the entire file.   
iv. Broadcasting Rules (Advanced):

 While basic broadcasting is common, the full set of broadcasting rules in NumPy can be quite sophisticated, enabling complex operations between arrays of different shapes. Understanding these rules is essential for writing efficient and concise code for advanced array manipulations.   
v. Universal Functions (ufuncs):

 NumPy provides a wide range of universal functions (ufuncs) that perform element-wise operations on arrays. Beyond basic arithmetic, ufuncs include trigonometric functions, exponential functions, logarithmic functions, and more. They are highly optimized for performance and can be customized.   
vi. Linear Algebra:

 NumPy offers a comprehensive set of functions for linear algebra operations, including matrix multiplication, inversion, eigenvalue decomposition, singular value decomposition, and more. These functions are essential for scientific computing and data analysis.   
vii. Fourier Transforms:

 NumPy provides functions for performing fast Fourier transforms (FFTs), which are widely used in signal processing, image analysis, and other scientific fields.   
viii. Random Number Generation:

 NumPy includes a powerful random number generation module that allows you to create arrays of random numbers from various probability distributions. This is essential for simulations, statistical modeling, and machine learning.   
ix. Array Broadcasting and Reduction:

 NumPy's broadcasting rules, combined with reduction operations (like sum, mean, max), allow for efficient calculations across different dimensions of arrays. This is useful for tasks like calculating the mean of each row or column in a 2D array.   
x. Custom ufuncs:

 For very specific performance needs, it's possible to create your own universal functions (ufuncs) using NumPy's C API. This allows for even greater control and optimization of array operations.   
xi. Numba Integration:

 NumPy can be further accelerated using Numba, a just-in-time compiler for Python. Numba can often compile NumPy code to machine code, resulting in significant performance improvements.   
xii. Cython Integration:

 For even finer-grained control and maximum performance, NumPy can be integrated with Cython, a language that allows you to write C extensions for Python. This is useful for highly specialized applications where performance is critical.   
These are just some of the advanced features offered by NumPy. Mastering these features can significantly enhance your ability to work with numerical data in Python and tackle complex scientific computing and data analysis problems.

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

# Sample DataFrame (replace with your data)
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 22, 35, 27],
        'Salary': [50000, 75000, 45000, 90000, 60000]}
df = pd.DataFrame(data)

# Threshold value (replace with your desired threshold)
salary_threshold = 60000

# Method 1: Boolean indexing (most common and efficient)
filtered_df = df[df['Salary'] > salary_threshold]

# Method 2: Using .loc with boolean indexing (more explicit)
filtered_df_loc = df.loc[df['Salary'] > salary_threshold]

# Method 3: Using .query() (more readable for complex conditions)
filtered_df_query = df.query('Salary > @salary_threshold') #The @ allows you to use variables in the query string.

# Printing the results
print("Original DataFrame:\n", df)
print("\nFiltered DataFrame (Method 1):\n", filtered_df)
print("\nFiltered DataFrame (Method 2):\n", filtered_df_loc)
print("\nFiltered DataFrame (Method 3):\n", filtered_df_query)


# Example with multiple conditions:
age_threshold = 25
filtered_multiple = df[(df['Salary'] > salary_threshold) & (df['Age'] > age_threshold)]
print("\nFiltered DataFrame (Multiple Conditions):\n", filtered_multiple)

# Example with string column filtering:
city_data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'City': ['New York', 'London', 'Paris', 'New York', 'London']}
city_df = pd.DataFrame(city_data)

city_filter = city_df[city_df['City'] == 'New York']
print("\nFiltered City DataFrame:\n", city_filter)
```

```python
#11. Perform matrix multiplication using NumPy

import numpy as np

# Example 1: Basic matrix multiplication

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

result_matrix = np.dot(matrix1, matrix2)  # Traditional way
# OR, more concisely (and preferred in modern Python):
result_matrix_modern = matrix1 @ matrix2       # Using the @ operator

print("Matrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)
print("Result of matrix multiplication (np.dot):\n", result_matrix)
print("Result of matrix multiplication (@ operator):\n", result_matrix_modern)



# Example 2: Matrix multiplication with different dimensions (important!)
matrix3 = np.array([[1, 2, 3], [4, 5, 6]])  # 2x3 matrix
matrix4 = np.array([[7, 8], [9, 10], [11, 12]]) # 3x2 matrix

result_matrix2 = matrix3 @ matrix4 # np.dot(matrix3, matrix4) also works

print("\nMatrix 3:\n", matrix3)
print("Matrix 4:\n", matrix4)
print("Result of matrix multiplication (different dimensions):\n", result_matrix2)


# Example 3: Multiplying a matrix by a vector
matrix5 = np.array([[1, 2], [3, 4]])
vector1 = np.array([5, 6])

result_vector = matrix5 @ vector1 # np.dot(matrix5, vector1) also works

print("\nMatrix 5:\n", matrix5)
print("Vector 1:\n", vector1)
print("Result of matrix multiplication (matrix x vector):\n", result_vector)

# Important Note: Element-wise multiplication is different.
element_wise_product = matrix1 * matrix2
print("\nElement-wise product of matrix1 and matrix2:\n", element_wise_product)
```

```python
#12. Use Pandas to load a CSV file and display its first 5 rows.

import pandas as pd

# Method 1: Most common and recommended

# Replace 'your_file.csv' with the actual path to your CSV file
try:
    df = pd.read_csv('your_file.csv')
    print(df.head())  # Displays the first 5 rows (default)

except FileNotFoundError:
    print(f"Error: File 'your_file.csv' not found. Please check the file path.")
except Exception as e:
    print(f"An error occurred while reading the CSV: {e}")


# Method 2: Using a relative path (if the CSV is in the same directory)

# If your CSV file is in the same directory as your Python script, you can use a relative path:
try:
    df_relative = pd.read_csv('./your_file.csv')  # ./ means current directory
    print("\nDataFrame from relative path:")
    print(df_relative.head())
except FileNotFoundError:
    print(f"Error: File 'your_file.csv' not found in the current directory.")
except Exception as e:
    print(f"An error occurred while reading the CSV: {e}")


# Method 3: Specifying the number of rows to read (useful for very large files)

# If you have a huge CSV and only want to read a few rows:
try:
    df_small = pd.read_csv('your_file.csv', nrows=10) #Reads only the first 10 rows.
    print("\nFirst 10 rows of the CSV:")
    print(df_small)
except FileNotFoundError:
    print(f"Error: File 'your_file.csv' not found.")
except Exception as e:
    print(f"An error occurred while reading the CSV: {e}")
```

```python
#13. Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go
import numpy as np  # For generating sample data (optional)

# 1. Sample Data (Replace with your actual data)
# Option 1: Use your existing data (e.g., from Pandas DataFrame columns)
# x = df['Column1']
# y = df['Column2']
# z = df['Column3']

# Option 2: Generate sample data (for demonstration)
np.random.seed(42)  # for reproducibility
n_points = 50
x = np.random.normal(0, 1, n_points)
y = np.random.normal(0, 1, n_points)
z = np.random.normal(0, 1, n_points)

# 2. Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')]) #mode = 'markers' makes it a scatter plot

# 3. Customize the layout (optional but recommended)
fig.update_layout(
    title="3D Scatter Plot",  # Replace with your title
    scene=dict(
        xaxis_title="X-axis Label",  # Replace with your x-axis label
        yaxis_title="Y-axis Label",  # Replace with your y-axis label
        zaxis_title="Z-axis Label",  # Replace with your z-axis label
        # Add other scene customization options here if needed
    ),
    margin=dict(l=0, r=0, b=0, t=50) #Adjust margins
)

# Optional: Customize marker appearance
fig.update_traces(marker=dict(size=6, color='blue', opacity=0.8)) # Change size, color and opacity of the markers

# Optional: Add color scale
# fig.update_traces(marker=dict(size=6, color=z, colorscale='Viridis', opacity=0.8)) # Color based on z-value

# Display the plot
fig.show()


# Example with different marker styles and colors for different groups:
n_points_group1 = 30
x1 = np.random.normal(0, 1, n_points_group1)
y1 = np.random.normal(0, 1, n_points_group1)
z1 = np.random.normal(0, 1, n_points_group1)

n_points_group2 = 20
x2 = np.random.normal(2, 1, n_points_group2)  # Different mean for group 2
y2 = np.random.normal(2, 1, n_points_group2)
z2 = np.random.normal(2, 1, n_points_group2)

fig2 = go.Figure()
fig2.add_trace(go.Scatter3d(x=x1, y=y1, z=z1, mode='markers', marker=dict(size=6, color='blue'), name='Group 1'))
fig2.add_trace(go.Scatter3d(x=x2, y=y2, z=z2, mode='markers', marker=dict(size=6, color='red'), name='Group 2'))

fig2.update_layout(title="3D Scatter Plot with Groups")
fig2.show()
```
